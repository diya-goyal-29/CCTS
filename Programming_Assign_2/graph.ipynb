```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bto_avg_time = [32,141,201, 381, 443]
mvto_un_avg_time = [2.5229, 27.7918, 17.4877, 51.1601, 26.6217]
mvto_gc_avg_time = [0.7216, 7.8349,9.0773, 104,143]
mvto_bn_avg_time = [6,10,15,56,170]

bto_avg_abort = [31,93,143, 227, 231]
mvto_un_avg_abort = [0,6,5,11,8]
mvto_gc_avg_abort = [0,3, 2, 21,25]
mvto_bn_avg_abort = [5,9,19,40,45]

x = [10,20,30,40,50]
```

```python
plt.scatter(x,bto_avg_time, c = 'red', marker='.', label = "BTO")
plt.scatter(x,mvto_un_avg_time, c='blue', marker='x', label = 'MVTO unbounded')
plt.scatter(x,mvto_gc_avg_time, c = 'green', marker='o', label = 'MVTO unbounded garbace collection')
plt.scatter(x,mvto_bn_avg_time, c = 'yellow', marker= '+', label = "MVTO bounded")
plt.plot(x,mvto_un_avg_time, c = 'blue')
plt.plot(x,mvto_gc_avg_time, c = 'green')
plt.plot(x,mvto_bn_avg_time, c = 'yellow')
plt.plot(x,bto_avg_time, c = 'red')
plt.xlabel("Num of threads")
plt.ylabel("Avg. Commit Delay")
plt.legend()
plt.show()
```

```python
plt.scatter(x,bto_avg_abort, c = 'red', marker='.', label = "BTO")
plt.scatter(x,mvto_un_avg_abort, c='blue', marker='x', label = 'MVTO unbounded')
plt.scatter(x,mvto_gc_avg_abort, c = 'green', marker='o', label = 'MVTO unbounded garbace collection')
plt.scatter(x,mvto_bn_avg_abort, c = 'yellow', marker= '+', label = "MVTO bounded")
plt.plot(x,mvto_un_avg_abort, c = 'blue')
plt.plot(x,mvto_gc_avg_abort, c = 'green')
plt.plot(x,mvto_bn_avg_abort, c = 'yellow')
plt.plot(x,bto_avg_abort, c = 'red')
plt.xlabel("Num of threads")
plt.ylabel("Avg. Aborts")
plt.legend()
plt.show()
```
